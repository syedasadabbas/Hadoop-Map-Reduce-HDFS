```python
#!/usr/bin/env python3

from operator import itemgetter
from itertools import groupby

# Function to parse input lines and generate key-value pairs
def parse_input(line):
    parts = line.strip().split('\t')
    uid = parts[0]
    if len(parts) >= 5:
        title, _, _, year, rating = parts[:5]
    else:
        title, year, rating = "", "", ""
    return uid, (title, rating, year)  # Include year in the tuple


# Function to format the output
def format_output(year, uid, data):
    title, rating, _ = data  # Unpack year from the data tuple
    appearance_count = len(rating)
    if appearance_count > 0:
        rating_str = ','.join([r[1] for r in rating])
        return f"{year}\t{title}\t{rating_str}\t{'[' + '1,' * (appearance_count-1) + '1]'}"
    else:
        return f"{year}\t{title}\t{uid}\t[1]"

def main():
    # Read input lines from mapout.txt
    with open('mapout.txt', 'r') as f:
        lines = f.readlines()

    # Count the occurrences of each movie ID
    movie_id_counts = {}
    for line in lines:
        movie_id = line.strip().split('\t')[0]
        movie_id_counts[movie_id] = movie_id_counts.get(movie_id, 0) + 1

    # Group input lines by movie ID
    grouped_lines = groupby(lines, key=lambda x: x.strip().split('\t')[0])

    # Process each group
    with open('comout.txt', 'w') as outfile:
        for uid, group in grouped_lines:
            # Initialize variables
            title = ''
            rating = []
            year = ''

            # Check if the movie appears only once
            if movie_id_counts.get(uid, 0) == 1:
                title, _, year, rating = parse_input(next(group))
                # Write output for the movie appearing only once
                output_line = format_output(year, uid, (title, [(title, '1')], year))
                outfile.write(output_line + '\n')
            else:
                # Iterate over group items
                for line in group:
                    parsed_data = parse_input(line)
                    title = parsed_data[1][0]
                    rating.append(parsed_data[1][1:])
                    year = parsed_data[1][2]  # Extract year

                # Format and write output
                output_line = format_output(year, uid, (title, rating, year))
                outfile.write(output_line + '\n')

if __name__ == "__main__":
    main()
```
